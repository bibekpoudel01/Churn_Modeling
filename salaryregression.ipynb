```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler , LabelEncoder, OneHotEncoder
import numpy as np
import pickle
```

```python
df=pd.read_csv('Churn_Modelling.csv')
```

```python
df=df.drop(columns=['RowNumber','CustomerId','Surname'])
```

```python
df.head()
```

```python
label_encoder=LabelEncoder()
df['Gender']=label_encoder.fit_transform(df['Gender'])
```

```python
ohe_encoder=OneHotEncoder(handle_unknown='ignore')
geo_encoded=ohe_encoder.fit_transform(df[['Geography']]).toarray()
```

```python
geo_encoded_df=pd.DataFrame(geo_encoded,columns=ohe_encoder.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
df=pd.concat([df.drop('Geography',axis=1),geo_encoded_df],axis=1)
```

```python
df.sample(4)
```

```python
##Split The data into feature and target
X=df.drop('EstimatedSalary',axis=1)
y=df['EstimatedSalary']
```

```python
with open('label_encoder_gender.pkl','wb')as file:
    pickle.dump(label_encoder,file)
with open('onehot_encoder_geo.pkl','wb')as file:
    pickle.dump(ohe_encoder,file)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
with open('scaler.pkl','wb')as file:
    pickle.dump(scaler,file)
```

```python
##ANN Regression Problem Statement
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential([
    Dense(64,activation='relu',input_shape=(X_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1)##OUtput LAyer for Regression ##It automatically give linear actiavtion function for regression 
    
    
    
    
    
    
])
```

```python
##Compile This model
model.compile(optimizer='adam',loss='mean_absolute_error',metrics=['mae'] )
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
log_dir="regressionlogs/fit/"+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
early_stopping=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
history=model.fit(X_train,y_train,
                  validation_data=(X_test,y_test),
                  epochs=100,
                  callbacks=[early_stopping,tensorboard])
```

```python
##Evaluate The model on the test data
test_loss,test_mae=model.evaluate(X_test,y_test)
print("Test MAE",test_mae)
```
